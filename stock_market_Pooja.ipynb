```python
import os
import requests
import gradio as gr
import openai
import matplotlib.pyplot as plt
from io import BytesIO
import base64

# 🔐 Load secrets from environment
client = openai.OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
STOCK_API_KEY = os.getenv("STOCK_API_KEY")

# 🔍 Improved stock symbol fetch with alias support
def get_stock_symbol(company_name):
    aliases = {
        "facebook": "META",
        "google": "GOOGL",
        "tesla": "TSLA",
        "apple": "AAPL",
        "amazon": "AMZN"
    }

    name_lower = company_name.lower()
    if name_lower in aliases:
        return aliases[name_lower]

    url = 'https://www.alphavantage.co/query'
    params = {
        'function': 'SYMBOL_SEARCH',
        'keywords': company_name,
        'apikey': STOCK_API_KEY
    }
    r = requests.get(url, params=params).json()
    try:
        best_match = next(
            match for match in r['bestMatches']
            if company_name.lower() in match['2. name'].lower()
        )
        return best_match['1. symbol']
    except (StopIteration, KeyError, IndexError):
        return None

# 📈 Get real-time stock price and last trading date
def get_stock_quote(symbol):
    url = 'https://www.alphavantage.co/query'
    params = {
        'function': 'GLOBAL_QUOTE',
        'symbol': symbol,
        'apikey': STOCK_API_KEY
    }
    r = requests.get(url, params=params).json()
    try:
        quote = r['Global Quote']
        return quote['05. price'], quote['07. latest trading day']
    except KeyError:
        return None, None

# 🧾 Get company financial info (overview)
def get_financial_overview(symbol):
    url = 'https://www.alphavantage.co/query'
    params = {
        'function': 'OVERVIEW',
        'symbol': symbol,
        'apikey': STOCK_API_KEY
    }
    r = requests.get(url, params=params).json()
    try:
        return {
            "P/E Ratio": r.get("PERatio", "N/A"),
            "EPS": r.get("EPS", "N/A"),
            "Market Cap": r.get("MarketCapitalization", "N/A"),
            "Description": r.get("Description", "No description available.")
        }
    except KeyError:
        return None

# 💬 Ask GPT for investment insight
def ask_gpt_summary(company_name, financial_data):
    prompt = (
        f"Provide a short summary and investment insight for {company_name}.\n"
        f"Here is some financial data:\n"
        f"P/E Ratio: {financial_data['P/E Ratio']}\n"
        f"EPS: {financial_data['EPS']}\n"
        f"Market Cap: {financial_data['Market Cap']}"
    )
    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a stock market assistant providing brief insights."},
                {"role": "user", "content": prompt}
            ],
            max_tokens=250,
            temperature=0.3
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        return f"GPT Error: {e}"

# 📊 Generate bar chart of financials
def safe_float(value):
    try:
        return float(value)
    except (ValueError, TypeError):
        return 0

def plot_financials(financial_data):
    labels = ['P/E Ratio', 'EPS', 'Market Cap ($B)']
    values = [
        safe_float(financial_data['P/E Ratio']),
        safe_float(financial_data['EPS']),
        safe_float(financial_data['Market Cap']) / 1e9
    ]
    plt.figure(figsize=(6, 3))
    bars = plt.barh(labels, values)
    for label, bar in zip(labels, bars):
        val = bar.get_width()
        text = f"${val:.2f}B" if 'Market Cap' in label else f"{val:.2f}"
        plt.text(val, bar.get_y() + bar.get_height() / 2, text, va='center')
    plt.xlabel('Value')
    plt.title('Company Financial Overview')
    plt.tight_layout()
    buf = BytesIO()
    plt.savefig(buf, format='png')
    plt.close()
    buf.seek(0)
    encoded = base64.b64encode(buf.read()).decode('utf-8')
    return f"data:image/png;base64,{encoded}"

# 🎯 Main function
def stock_insight(company_name):
    symbol = get_stock_symbol(company_name)
    if not symbol:
        return f"❌ Could not find stock symbol for '{company_name}'.", "", "", "", "", "", None

    price, last_trade = get_stock_quote(symbol)
    if not price:
        return f"❌ Could not fetch price for symbol {symbol}.", "", "", "", "", "", None

    overview = get_financial_overview(symbol)
    if not overview:
        return f"❌ Could not fetch financial data for {symbol}.", "", "", "", "", "", None

    summary = ask_gpt_summary(company_name, overview)
    chart = plot_financials(overview)

    return (
        f"✅ Symbol: {symbol}",
        f"💲 Price: ${price}",
        f"🗓️ Last Trading Day: {last_trade}",
        f"📊 P/E Ratio: {overview['P/E Ratio']}\nEPS: {overview['EPS']}\nMarket Cap: ${overview['Market Cap']}",
        f"📝 Company Summary: {overview['Description'][:300]}...",
        f"🤖 GPT Insight: {summary}",
        chart
    )

# 🖥️ Gradio Interface with Blocks
with gr.Blocks() as demo:
    gr.Markdown("# 📈 Stock Market Assistant")
    with gr.Row():
        company_input = gr.Textbox(label="Enter Company Name")
        search_button = gr.Button("🔍 Search")

    with gr.Row():
        with gr.Column():
            symbol_output = gr.Textbox(label="Symbol")
            price_output = gr.Textbox(label="Stock Price")
            trade_output = gr.Textbox(label="Last Trading Day")
            financial_output = gr.Textbox(label="Financial Overview")
        with gr.Column():
            chart_output = gr.Image(label="📊 Financial Chart")

    company_summary = gr.Textbox(label="Company Summary")
    gpt_output = gr.Textbox(label="AI Insight")

    search_button.click(stock_insight, inputs=company_input, outputs=[
        symbol_output, price_output, trade_output,
        financial_output, company_summary, gpt_output, chart_output
    ])

demo.launch()
```
